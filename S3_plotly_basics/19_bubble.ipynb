```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
from umaplotly import uma

pio.templates.default = uma
```

```python
df = pd.read_csv(r'..\source\Plotly-Dashboards-with-Dash\Data\mpg.csv')
df
```

```python
def scaling_bubble_size(size, desired_size):
    return 2.0 * max(size) / (desired_size**2)


data = [
    go.Scatter(
        x=df.horsepower,
        y=df.mpg,
        text=df.name,
        mode="markers",
        marker=dict(
            size=df.weight,
            sizeref=scaling_bubble_size(df.weight, 8),
            color=df.cylinders,
            showscale=True,
        ),
    )
]

layout = go.Layout(title="Bubble Chart", hovermode="closest")

go.Figure(data=data, layout=layout)
```
