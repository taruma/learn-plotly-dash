# MULTIPLE HEAT MAPS WITH SUBPLOTS

```python
from plotly.subplots import make_subplots

df1 = pd.read_csv(
    r'..\source\Plotly-Dashboards-with-Dash\Data\2010SitkaAK.csv')
df2 = pd.read_csv(
    r'..\source\Plotly-Dashboards-with-Dash\Data\2010SantaBarbaraCA.csv')
df3 = pd.read_csv(r'..\source\Plotly-Dashboards-with-Dash\Data\2010YumaAZ.csv')

titles = ['Sitka AK', 'SB CA', 'Yuma AZ']

traces = [
    go.Heatmap(
        x=df.DAY, y=df.LST_TIME, z=df.T_HR_AVG, colorscale='jet', zmin=5, zmax=40, name=title
    )
    for df, title in zip([df1, df2, df3], titles)
]


fig = make_subplots(
    rows=1, cols=3, subplot_titles=titles, shared_yaxes=True
)

for i, trace in enumerate(traces, 1):
    fig.add_trace(trace, 1, i)

fig.update_layout(title='Temp for 3 Cities')
```
