```python
#######
# Objective: Create a bubble chart that compares three other features
# from the mpg.csv dataset. Fields include: 'mpg', 'cylinders', 'displacement'
# 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name'
######

# Perform imports here:

import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
from umaplotly import uma

pio.templates.default = uma

```

```python

# create a DataFrame from the .csv file:
df = pd.read_csv(r'..\source\Plotly-Dashboards-with-Dash\Data\mpg.csv')
df
```

```python
# create data by choosing fields for x, y and marker size attributes


def scaling_bubble_size(size, desired_size):
    return 2.0 * max(size) / (desired_size**2)


data = [
    go.Scatter(
        x=df.displacement,
        y=df.acceleration,
        text=df.name,
        mode="markers",
        marker=dict(
            size=df.weight,
            sizeref=scaling_bubble_size(df.weight, 8),
            # color=df.mpg,
            # showscale=True,
        ),
    )
]

layout = go.Layout(title="Bubble Chart", hovermode="closest")

go.Figure(data=data, layout=layout)
```
