```python
#######
# Objective: Create a histogram that plots the 'length' field
# from the Abalone dataset (../data/abalone.csv).
# Set the range from 0 to 1, with a bin size of 0.02
######

# Perform imports here:
import plotly.graph_objects as go
import pandas as pd
import plotly.io as pio
from umaplotly import uma_template

pio.templates.default = uma_template

# create a DataFrame from the .csv file:

df = pd.read_csv(r'..\source\Plotly-Dashboards-with-Dash\Data\abalone.csv')
df
```

```python

# create a data variable:

data = [
    go.Histogram(x=df.length, xbins=dict(start=0, end=1, size=0.02))
]

# add a layout

layout = go.Layout(title='Histogram Length Petal')

# create a fig from data & layout, and plot the fig

go.Figure(data, layout)
```
