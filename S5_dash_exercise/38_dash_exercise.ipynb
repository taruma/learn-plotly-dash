```python
from dash import html, dcc
from jupyter_dash import JupyterDash
import plotly.io as pio
import plotly.graph_objects as go

pio.templates.default = 'none'
app = JupyterDash()
```

```python
#######
# Objective: build a dashboard that imports OldFaithful.csv
# from the data directory, and displays a scatterplot.
# The field names are:
# 'D' = date of recordings in month (in August),
# 'X' = duration of the current eruption in minutes (to nearest 0.1 minute),
# 'Y' = waiting time until the next eruption in minutes (to nearest minute).
######

# Perform imports here:

import pandas as pd

df = pd.read_csv(r'..\source\Plotly-Dashboards-with-Dash\Data\OldFaithful.csv')
df
```

```python
app = JupyterDash()
data = [go.Scatter(x=df.X, y=df.Y, mode="markers")]

layout = go.Layout(
    title="Old Faithful Eruption Intervals v Durations",
    xaxis=dict(title="Duration of eruption (minutes)"),
    yaxis=dict(title="Interval to next eruption (minutes)"),
    hovermode='closest'
)

fig = go.Figure(data, layout)
fig

```

```python
# Launch the application:

app.layout = html.Div(children=[
    dcc.Graph(id='faithful', figure=fig)
])

# Add the server clause:
```

```python
app.run_server(mode='external')
```
